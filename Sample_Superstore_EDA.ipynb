## Cek outliers

### 1. Sales

```python
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

outlier_sales = df[
    (df['Sales'] < lower_bound) |
    (df['Sales'] > upper_bound)
]

print("Jumlah Outlier Sales:", len(outlier_sales))
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

outlier_sales.head()
```

```python
# Produk apa yang menyebabkan Sales menjadi outlier?
# Lihat transaksi Sales terbesar

display(
    outlier_sales[
        ['Product Name',
         'Category',
         'Sales',
         'Profit',
         'Discount']
    ]
    .sort_values('Sales', ascending=False)
    .head(10)
)
```

### 2. Profit

```python
Q1 = df['Profit'].quantile(0.25)
Q3 = df['Profit'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

outlier_profit = df[
    (df['Profit'] < lower_bound) |
    (df['Profit'] > upper_bound)
]

print("Jumlah Outlier Profit:", len(outlier_profit))
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

outlier_profit.head()
```

```python
# Produk apa yang paling menguntungkan dan paling merugikan?
# Cari profit tertinggi

display(
    outlier_profit[
        ['Product Name',
         'Category',
         'Sales',
         'Profit',
         'Discount']
    ]
    .sort_values('Profit', ascending=False)
    .head(10)
)
```

```python
# Cari kerugian terbesar

display(
    outlier_profit[
        ['Product Name',
         'Category',
         'Sales',
         'Profit',
         'Discount']
    ]
    .sort_values('Profit')
    .head(10)
)
```

### 3. Discount

```python
Q1 = df['Discount'].quantile(0.25)
Q3 = df['Discount'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

outlier_discount = df[
    (df['Discount'] < lower_bound) |
    (df['Discount'] > upper_bound)
]

print("Jumlah Outlier Discount:", len(outlier_discount))
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

outlier_discount.head()
```

```python
# Apakah diskon tinggi menghasilkan profit atau justru rugi?
# Cari transaksi dengan diskon tertinggi

display(
    outlier_discount[
        ['Product Name',
         'Sales',
         'Profit',
         'Discount']
    ]
    .sort_values('Discount', ascending=False)
    .head(10)
)
```

### 4. Quantity

```python
Q1 = df['Quantity'].quantile(0.25)
Q3 = df['Quantity'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

outlier_quantity = df[
    (df['Quantity'] < lower_bound) |
    (df['Quantity'] > upper_bound)
]

print("Jumlah Outlier Quantity:", len(outlier_quantity))
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

outlier_quantity.head()
```

```python
# Produk apa yang dibeli dalam jumlah sangat besar?

display(
    outlier_quantity[
        ['Product Name',
         'Sales',
         'Profit',
         'Quantity']
    ]
    .sort_values('Quantity', ascending=False)
    .head(10)
)
```

## Grouping & Sorting

### 1. Discount vs Profit

```python
df.groupby('Discount')[['Sales', 'Profit']].mean().sort_values(by='Profit', ascending=False)
```

### 2. Profit by Category 

```python
profit_category = (
    df.groupby('Category')
      .agg({
          'Sales':'sum',
          'Profit':'sum'
      })
      .sort_values('Profit', ascending=False)
)

display(profit_category)
```

### 3. Profit by Sub-Category

```python
profit_subcategory = (
    df.groupby('Sub-Category')
      .agg({
          'Sales':'sum',
          'Profit':'sum'
      })
      .sort_values('Profit')
)

display(profit_subcategory)
```

### 4. Profit by Region

```python
profit_region = (
    df.groupby('Region')
      .agg({
          'Sales':'sum',
          'Profit':'sum'
      })
      .sort_values('Profit', ascending=False)
)

display(profit_region)
```

### 5. Discount Impact vs Profit

```python
discount_profit = (
    df.groupby('Discount Impact')
      .agg({
          'Sales':'sum',
          'Profit':'mean'
      })
      .sort_values('Profit', ascending=False)
)

display(discount_profit)
```

### 6. Discount Impact vs Profit Ratio

```python
discount_ratio = (
    df.groupby('Discount Impact')
      .agg({
          'Profit Ratio':'mean'
      })
      .sort_values('Profit Ratio', ascending=False)
)

display(discount_ratio)
```

### 7. Profit vs Loss

```python
profit_status = (
    df.groupby('Profit Status')
      .agg({
          'Sales':'sum',
          'Profit':'sum',
          'Discount':'mean'
      })
)

display(profit_status)
```

### 8. Sales Category vs Profit

```python
sales_category_profit = (
    df.groupby('Sales Category')
      .agg({
          'Sales':'mean',
          'Profit':'mean',
          'Profit Ratio':'mean'
      })
)

display(sales_category_profit)
```
